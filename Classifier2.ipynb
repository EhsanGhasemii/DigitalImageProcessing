```python
# 0) importing our needed libraries
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
# 1) Creating our dataset
n_samples = 600
X, Y = datasets.make_circles(n_samples=n_samples,
                             noise=0.17,
                             random_state=11,
                             factor=0.1
                             )
x_data = torch.Tensor(X)
y_data = torch.Tensor(Y.reshape(n_samples, 1))

def plot_our_dataset(X, Y):
    plt.scatter(X[Y==0, 0], X[Y==0, 1])
    plt.scatter(X[Y==1, 0], X[Y==1, 1])
```

```python
plot_our_dataset(X, Y)
```

```python
# 2) Building our model 
class Classifier2(nn.Module):
    def __init__(self, input_size, H1, output_size):
        super().__init__()
        self.linear1 = nn.Linear(input_size, H1)
        self.linear2 = nn.Linear(H1, output_size)
#         self.linear3 = nn.Linear(H2, output_size)
        
    def forward(self, x):
        x = torch.sigmoid(self.linear1(x))
        x = torch.sigmoid(self.linear2(x))
#         x = torch.sigmoid(self.linear3(x)1)
        return x
    
    def predict(self, x):
        pred = self.forward(x)
        pred[pred>=0.5] = 1
        pred[pred<0.5] = 0
        return pred
    
model = Classifier2(2, 4, 1)
print(model)
```

```python
criterion = nn.BCELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.1)
```

```python
# 3) Training our model 
epochs = 1024
losses = []

for e in range(epochs):
    pred = model(x_data)
    loss = criterion(pred, y_data)
    
    print("epoch : ", e, " loss : ", loss.item())
    losses.append(loss.item())
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

```python
# Showing the losses redundent
plt.plot(range(epochs), losses)
```

```python
# 4) Plotting the result
def meshgrid_plot(X, Y):
    my_resolution = 200
    x_span = np.linspace(np.min(X[:, 0])-0.25,
                         np.max(X[:, 0])+0.25,
                         num=my_resolution)
    y_span = np.linspace(np.min(X[:, 1])-0.25,
                         np.max(X[:, 1])+0.25,
                         num=my_resolution)
    x_axis, y_axis = np.meshgrid(x_span, y_span)
    my_grid = torch.Tensor(np.c_[x_axis.ravel(), y_axis.ravel()])
    prediction_of_my_grid = model.predict(my_grid)
    prediction_of_my_grid = prediction_of_my_grid.view(x_axis.shape)
    prediction_of_my_grid = prediction_of_my_grid.detach().numpy()
    plt.contourf(x_axis, y_axis, prediction_of_my_grid)
```

```python
meshgrid_plot(X, Y)
plot_our_dataset(X, Y)
```

```python
# 5) Testting our model 
point1 = torch.Tensor([0.123, 0.456])
point1_predict = model.predict(point1)
print('Prediction of this point is : ', point1_predict.item())
meshgrid_plot(X, Y)
plt.plot(point1.numpy()[0], point1.numpy()[1], 'ro')
```
