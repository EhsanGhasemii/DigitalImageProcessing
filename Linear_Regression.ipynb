# Linear Regression


```python
# 0) importing our needed library 
import torch
import numpy as np
import matplotlib.pyplot as plt
import torch.nn as nn
```

```python
# 1) Creating our dataset
X = torch.randn(size=(100, 1))*10
Y = 3*X + 4 + torch.randn(size=X.shape)*3
plt.plot(X.numpy(), Y.numpy(), 'o')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Our dataset")
plt.show()
```

```python
# 2) Building our model 
class Linear_Regression(nn.Module):
    def __init__(self, input_size, output_size):
        super().__init__()
        self.linear = nn.Linear(input_size, output_size)
    def forward(self, x):
        pred = self.linear(x)
        return pred
```

```python
# 2+) Define our model 
model = Linear_Regression(1, 1)
print(model)
[w, b] = model.parameters()
w = w.item()
b = b.item()
print(w, b)
plt.plot(X, Y, 'o')
input_x = np.array([-30, 30])
output_y = w * input_x + b
plt.plot(input_x, output_y)
plt.show()
```

```python
criterion = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)
```

```python
# 3) Train our model 
epochs = 1000
losses = []

for e in range(epochs):
    preds = model(X)
    loss = criterion(preds, Y)
    losses.append(loss.item())
    
    print('epoch : ', e, ', loss : ', loss.item())
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

```python
plt.plot(range(epochs), losses)
plt.xlabel('epoch')
plt.ylabel('loss')
```
